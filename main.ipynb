<h3>Null Value Detection & Cleanup</h3>

```python
null_count = df.isnull().sum()
null_pct= ((null_count/len(df))* 100).round(1)

print(null_pct)
```

```python
df_dropped = df.dropna()
print(f"Original  : {df.shape[0]} rows")
print(f"After dropna: {df_dropped.shape[0]} rows")
print(f"Lost rows : {df.shape[0] - df_dropped.shape[0]}")
```

```python
cols_to_fill=[
    "seeks_career_counseling",
    "self_learning_hours_per_week",
    "social_media_hrs_per_day",
    "sleep_hours"
]
print("Before:")
for col in cols_to_fill:
    print(f"    {col} null count: ", df[col].isnull().sum())
    print(f"    Median: {df[col].median()}")
for col in cols_to_fill:
    df[col] = df[col].fillna(df[col].median())
print("After:")
for col in cols_to_fill:
    print(f"    {col} null count: ", df[col].isnull().sum())
    print(f"    Median: {df[col].median()}")
print("Total nulls: ",df.isnull().sum().sum())
print("Columns and rows: ",df.shape)
```

<h3>Outlier Detection</h3>

```python
numerical_cols = df.select_dtypes(include="number").columns
for col in numerical_cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR = Q3-Q1
    lower = Q1-1.5*IQR
    upper = Q3+1.5 * IQR
    count = df[(df[col] < lower) | (df[col] > upper)].shape[0]
    if count >0:
        print(f"{col}: {count} outliers  (lower={lower:.2f}, upper={upper:.2f})")
short_names = {col: col[:8] for col in numerical_cols}
plt.figure(figsize=(16, 6))
sns.boxplot(data=df[numerical_cols].rename(columns=short_names))
plt.title("Outlier Detection")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.tight_layout()
plt.show()
```

```python
cols_to_remove_outliers=[
    "daily_ai_tool_usage_hrs",
    "internship_experience",
    "weekly_job_application_count",
    "self_learning_hours_per_week",
    "skill_development_courses_taken",
    "social_media_hrs_per_day"
]
print("Before",df.shape)
for col in cols_to_remove_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
print("After",df.shape)
plt.figure(figsize=(16, 6))
sns.boxplot(data=df[numerical_cols].rename(columns=short_names))
plt.title("Outlier Detection")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.tight_layout()
plt.show()
```

<h3>Visualizing with Seaborn</h3>

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y="primary_ai_tools_used",order=df["primary_ai_tools_used"].value_counts().index)
plt.title('AI Tools Used')
plt.xlabel('Number of Students')
plt.ylabel('AI Tool')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data=df, x='daily_ai_tool_usage_hrs', ax=axes[0],bins=30, kde=True)
axes[0].set_title('Distribution of Daily AI Tool Usage (Hours)')
axes[0].set_xlabel('Hours per day')
axes[0].set_ylabel('Count')

sns.scatterplot(data=df, x='ai_dependency_score', y='overall_career_readiness_score', ax=axes[1],hue='uses_ai_for_assignments', alpha=0.5,)
axes[1].set_title('AI Dependency vs Career Readiness')
axes[1].set_xlabel('AI Dependency Score')
axes[1].set_ylabel('Career Readiness Score')

plt.tight_layout()
plt.show()
```

```python
fear_of_job_loss = df.groupby(['fear_of_job_loss_to_ai', 'stream']).size().reset_index(name='Count')
sns.barplot(data=fear_of_job_loss, x='fear_of_job_loss_to_ai', y='Count', hue='stream')

plt.title('Fear of Job Loss to AI by Stream')
plt.xlabel('Fear Level (1-5 Scale)')
plt.ylabel('Count')
```

```python
key_vars = ['daily_ai_tool_usage_hrs', 'ai_dependency_score', 'placement_anxiety_score',
            'career_clarity_score', 'overall_career_readiness_score', 'stress_level',
            'burnout_score', 'motivation_score', 'daily_study_hours', 'sleep_hours',
            'interview_anxiety_score', 'resume_confidence_score']
plt.figure(figsize=(14, 12))
corr_matrix = df[key_vars].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f',
            square=True, linewidths=0.5)
plt.title('Correlation Heatmap of Key Variables', fontsize=14)
plt.tight_layout()
plt.show()
```
